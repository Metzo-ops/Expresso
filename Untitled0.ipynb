```python
#Conversion d'une chaine en boolÃ©en
from sklearn.compose import make_column_selector
from sklearn.preprocessing import OneHotEncoder 
X = expresso.drop(["CHURN","REGION","TOP_PACK","TENURE"],axis = 1)
X.MRG = X["MRG"].fillna("NO")
# make_column_selector sert Ã  selectionner une colonne selon le type de donnÃ©es
cat=make_column_selector(dtype_include=object)
col_cat=cat(X)
tab_cat=X[col_cat]
tab_cat=OneHotEncoder().fit_transform(tab_cat).data
# X[["MRG"]] = OneHotEncoder().fit_transform(X[["MRG"]])


tab_cat=pd.DataFrame(tab_cat, columns=["MRG"])
X["MRG"]=tab_cat
X.head()

```

```python
X["MRG"]
```

```python
Y = expresso.CHURN
Y = Y.fillna(0.0)
# print(Y)
X.REGULARITY = X.REGULARITY.fillna(X.REGULARITY.mean())
X.isna().sum()
```

```python
X_test = expresso.drop(["CHURN","REGION","TOP_PACK","TENURE"],axis = 1)
X_test.MRG = X_test["MRG"].fillna("NO")
# make_column_selector sert Ã  selectionner une colonne selon le type de donnÃ©es
cat_1=make_column_selector(dtype_include=object)
col_cat_1=cat_1(X_test)
tab_cat_1=X_test[col_cat_1]
tab_cat_1=OneHotEncoder().fit_transform(tab_cat_1).data
# X[["MRG"]] = OneHotEncoder().fit_transform(X[["MRG"]])


tab_cat_1=pd.DataFrame(tab_cat_1, columns=["MRG"])
X_test["MRG"]=tab_cat_1
X_test.head()
X_test.REGULARITY = X_test.REGULARITY.fillna(X_test.REGULARITY.mean())
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

parameters ={"loss":["deviance"]}
mymodel = GradientBoostingClassifier(learning_rate=1,n_estimators=10,verbose=2,max_depth=5,warm_start=True)
grid = GridSearchCV(mymodel,parameters)
grid.fit(X,Y)
grid.best_estimator_
```

```python
from sklearn.metrics import accuracy_score
Y_pred = grid.predict(X)
print(accuracy_score(Y_pred, Y))
```

```python
Y_test_pred = grid.predict(X_test)
print(accuracy_score(Y_test_pred, Y))
```
